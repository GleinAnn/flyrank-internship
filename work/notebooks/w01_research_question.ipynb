Lane 2: Refresh / Content Opportunity Scoring

Possible research question:

> *Can historical content-performance data rank pages that should be reviewed first by combining their future decline risk with the likely performance impact of that decline?*

The proposed system will estimate two components for each page:
$$ PriorityScore = P(future decline) \times Potential Impact$$

The decline probability will be predicted from historical metrics, such as clicks, impressions, CTR, search position, and engagement. Potential impact will measure the amount of traffic or visibility that may be lost, for example through expected lost clicks. Pages with higher scores will be placed higher in the review queue because they are both more likely to decline and more costly to ignore.

Reason:

> Content teams usually have limited time and cannot review every page. Selecting pages based only on their current health may prioritize low-value declines while missing pages with greater traffic or visibility at risk. This project therefore aims to build a practical ranking system that helps content editors and SEO teams decide which pages should be investigated first.

## 3. Quick look at the data (2-3 real numbers)

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

!git clone https://github.com/GleinAnn/flyrank-internship.git

import pandas as pd

df = pd.read_csv("flyrank-internship/data/raw/content_refresh_anonymized.csv")

import numpy as np
import pandas as pd

example = df.copy()

# 1. Temporary decline signal
# Example: trend_pct = -35 means a decline signal of 0.35
example["decline_signal"] = (
    -example["trend_pct"].fillna(0) / 100
).clip(lower=0, upper=1)

# 2. Impact: visibility relative to all pages
# log1p reduces the effect of extremely large pages
example["impact_score"] = (
    np.log1p(example["impressions_90d"])
    .rank(pct=True)
)

# 3. Combined priority score
example["priority_score"] = (
    100
    * example["decline_signal"]
    * example["impact_score"]
)

# Keep pages showing some decline
declining_pages = example[
    (example["trend_pct"] < 0)
    & (example["impressions_90d"] > 0)
].copy()

# Select three representative pages:
# low, medium, and high visibility
impact_levels = [0.25, 0.50, 0.75]

selected_indices = [
    (declining_pages["impact_score"] - level).abs().idxmin()
    for level in impact_levels
]

sample_3 = (
    declining_pages.loc[
        selected_indices,
        [
            "content_id",
            "impressions_90d",
            "clicks_90d",
            "trend_pct",
            "decline_signal",
            "impact_score",
            "priority_score",
        ],
    ]
    .drop_duplicates()
    .sort_values("priority_score", ascending=False)
)

sample_3.round(3)
```
